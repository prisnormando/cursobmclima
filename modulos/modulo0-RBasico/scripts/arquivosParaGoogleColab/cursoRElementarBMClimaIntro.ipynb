Ao longo do material disponvel voc ter contato com os fundamentos da programao de computadores aplicada  anlise de dados utilizando a linguagem R. ### Conceitos fundamentais
* **Introduo ao R**: o que  o R e o que podemos fazer com ele

### Seu espao de trabalho
* **Instalando o R, RStudio, GIT e Colab**: instalao das ferramentas de trabalho e algumas alternativas teis

### Fundamentos
* **Variveis e constantes**: entenda o que permite uma linguagem de programao manupular dados inseridos (entrada), calcular (processamento) e entregar resultados (sada)
* **Operadores**: use o R como uma calculadora

* **Manipulao de strings**: trate caracteres alfanumricos e encontre insights dentro de textos com o R
* **Estruturas de dados**: como estruturar dados no R
* **Entradas e sadas (input/output)**: as formas de inserir dados no R e encontrar os resultados esperados
* **Dicionrios de Dados**: como associar um ou mais tipos de dados no R
* **Estruturas condicionais**: como utilizar o R para realizar tarefas complexas utilizando fluxos de processamento por meio de decises condicionais
* **Loops**: como utilizar o R para realizar tarefas complexas utilizando fluxos de processamento por meio de repetio de tarefas
